```python
import tkinter as tk
from random import randint

root = tk.Tk()
root.title("Number Guessing Game")

# Set the background color
root.configure(bg="#87CEEB")

random_number = randint(1, 10)
attempts = 0

# Create a frame to hold the title
title_frame = tk.Frame(root, bg="#cc9918", highlightbackground="#000000", highlightthickness=1)
title_frame.pack(pady=20)

# Create a label to display the title
title_label = tk.Label(title_frame, text="Number Guessing Game", font=("Arial", 20), bg="#cc9918", fg="#FFFFFF")
title_label.pack(pady=20)

# Create a frame to hold the entry field and label
entry_frame = tk.Frame(root, bg="#87CEEB")
entry_frame.pack(pady=20)

# Create a label to display the instructions
instruction_label = tk.Label(entry_frame, text="Guess a number between 1 and 10:", font=("Arial", 16), bg="#87CEEB", fg="#FFFFFF")
instruction_label.pack(side=tk.TOP)

# Create an entry field for the user to input their guess
guess_entry = tk.Entry(entry_frame, font=("Arial", 16), width=10)
guess_entry.pack(side=tk.TOP, pady=10)

# Create a button to submit the user's guess
submit_button = tk.Button(entry_frame, text="Submit", font=("Arial", 16), bg="#4CAF50", fg="#FFFFFF", command=lambda: check_guess())
submit_button.pack(side=tk.TOP, pady=10)

def check_guess():
    global attempts
    attempts += 1
    try:
        guess = int(guess_entry.get())
        if guess < random_number:
            result_label.config(text=f"Try again! Your guess ({guess}) is less than the number.")
        elif guess > random_number:
            result_label.config(text=f"Try again! Your guess ({guess}) is greater than the number.")
        else:
            result_label.config(text="Congratulations! You guessed the number.")
            submit_button.config(state="disabled")
    except ValueError:
        result_label.config(text="Invalid input. Please enter a number.")
    if attempts >= 5:
        result_label.config(text=f"You've run out of attempts! The number was {random_number}.")
        submit_button.config(state="disabled")

# Create a label to display the result
result_label = tk.Label(root, text="", font=("Arial", 16), bg="#87CEEB", fg="#FFFFFF")
result_label.pack(pady=20)

root.geometry("500x400")
root.mainloop()
```
